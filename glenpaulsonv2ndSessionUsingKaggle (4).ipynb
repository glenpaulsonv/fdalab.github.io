```r
#Naming the Array
vec1 <- c(2,4,6)
vec2 <- c(11,12,13,14,15,16)
column.names <- c("COLA","COLB","COLC")
row.names <- c("ROWA","ROWB","ROWC")
matrix.names <- c("MatA","MatB")

res <- array(c(vec1,vec2), dim = c(3,3,2), dimnames = list(column.names, row.names))
print(res)
```

```r
name <- c("Glen"," Paulson","Praveen","Bhargav","Rushik")
age <- c(20,54,18,18,18)
vehicle <- c("Harley","Enfield","Enfield","KTM","KTM")
g = data.frame(name,age,vehicle)
print(g)
#Adding columns in the data frame
g$sex = c('m','m','m','m','m')
print(g)
#Adding Rows in the data frame
name <- c("Jaleena")
age <- c(20)
vehicle <- c("Lamborgini")
sex <- c('f')
print(g)
frame1 <- data.frame(name,age,vehicle,sex)
framefinal <- rbind(g,frame1)
print(framefinal)
```
